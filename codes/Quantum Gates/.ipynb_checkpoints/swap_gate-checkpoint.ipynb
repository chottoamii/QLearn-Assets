# Initialization

Importing necessary components and defining necessary functions just as before

```python
from qiskit import QuantumCircuit, transpile, Aer, execute
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import math 
```

```python
def get_result_from_backend(qc, shots=1000, simulator_name='qasm_simulator'):
    
    # Choose the backend (simulator or real quantum device)
    backend = Aer.get_backend(simulator_name)

    # Transpile the quantum circuit for the chosen backend
    transpiled_qc = transpile(qc, backend)
    
    # Execute the quantum circuit on the chosen backend
    result = execute(transpiled_qc, backend, shots=shots).result()
    
    return result
```

```python
def get_unitary_matrix(qc):
    result = get_result_from_backend(qc,simulator_name='unitary_simulator')
    return result.get_unitary(qc, decimals=3)
```

```python
def get_state_vector(qc):
    result = get_result_from_backend(qc,simulator_name='statevector_simulator')
    statevector = result.get_statevector()

    # Visualize the Bloch sphere representation of the final state
    return statevector
```

# Motivation

Normally, how do we swap the content of two variables?
The following code is used quite often, although there are smarter alternatives.

```python
a = 5
b = 3

t = a
a = b 
b = t 

print(a)
print(b)
```

But how do we do the same for two qubits? If we want to swap the states of two qubits, the suitable gate is SWAP gate.

# A Simple Demonstration

```python
# Create a quantum circuit with 2 qubits
qc = QuantumCircuit(2)

qc.h(0)
qc.swap(0, 1)

qc.measure_all()

qc.draw('mpl')
```

Let's plot the counts of observed states.

```python
result = get_result_from_backend(qc)

# Get the counts of the measurements
counts = result.get_counts()

# Plot the histogram
plot_histogram(counts)
```

Here, the output must be interpreted carefully.

First, recall that, in Qiskit, the lower qubit represents the MSB. And look the plot. For all of the observations, the LSB turns out to be |0>, however, the MSB fluctuates almost uniformly. That shows that although the Hadamard was originally applied on the upper qubit ( the LSB ), the states have been swapped by the SWAP gate.

# What's the matrix?

Just like before, let's verify the unitary matrix for SWAP gate in qiskit.

```python
qc = QuantumCircuit(2)

qc.swap(0,1)

print(get_unitary_matrix(qc))
```

# Let's Look at the Bloch Spheres

Observe how the state-vectors on the Bloch Spheres change as we gradually build the first circuit step by step.

```python
qc = QuantumCircuit(2)

plot_bloch_multivector(get_state_vector(qc))
```

```python
qc.h(0)

plot_bloch_multivector(get_state_vector(qc))
```

```python
qc.swap(0,1)

plot_bloch_multivector(get_state_vector(qc))
```

Look how the states swap between the two qubits. Impressive, isn't it?

# Can We Control?

Maybe this could be a separate lesson itself, but since it's related, let's look at it right here.

Just like Controlled NOT gate, we can also have controlled SWAP gate. But the difference is, for this, there is one control qubit and two target qubits.

When the control qubit is at |1>, the target qubits get swapped. 

Let's visualize in Bloch Sphere.

```python
qc = QuantumCircuit(3)

qc.x(0)
qc.x(1)
qc.h(1)
qc.h(2)
qc.cswap(0,1,2)

qc.draw('mpl')
```

```python
plot_bloch_multivector(get_state_vector(qc))
```

Please try to understand by yourself, what's happening here.

And what should happen, if we remove the Pauli-X gate from the qubit \(q0\)?
