# Observing the effect on one qubit

Since the idea of Pauli gates is almost similar for all the three gates, let us first explore the Pauli X gate in detail.

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile, Aer, execute
from qiskit.visualization import plot_histogram, plot_bloch_multivector
```

```python
# Create a quantum circuit with one qubit and one classical bit for measurement
qc = QuantumCircuit(1,1)

# Apply Pauli-X gate to the first qubit
qc.x(0)

# Measure the qubits
qc.measure([0],[0])  # measure qubit[0] and store the result into classical_bit[0]

qc.draw(output='mpl')
```

Now, since we have already explored how to run a circuit on quantum backend, let's write a function for this for our future use.

```python
def get_result_from_backend(qc, shots=1000, simulator_name='qasm_simulator'):
    
    # Choose the backend (simulator or real quantum device)
    backend = Aer.get_backend(simulator_name)

    # Transpile the quantum circuit for the chosen backend
    transpiled_qc = transpile(qc, backend)

    # Execute the quantum circuit on the chosen backend
    result = execute(transpiled_qc, backend, shots=shots).result()
    
    return result
    
```

```python
result = get_result_from_backend(qc, 1000)
    
# Get the counts of the measurements
counts = result.get_counts()

# Plot the histogram
plot_histogram(counts)
```

As expected, we obtain, only the state $|1\rangle$ as Pauli X-gate acts somewhat like a boolean NOT gate and switches the input from $|0\rangle$ to $|1\rangle$

# Are Pauli Gates self-invertible ?

This time, let's just avoid the math and do a hands-on experiment.

We will build a single quantum circuit to explore Pauli X, Y, Z gates on three different qubits. And to test self-invertibility, we will include each of them twice in rows. If the are indeed all self-invertible, the final measurement should only output the state $|000\rangle$

```python
qc = QuantumCircuit(3,3)

qc.x(0)
qc.x(0)

qc.y(1)
qc.y(1)

qc.z(2)
qc.z(2)

qc.measure_all()

qc.draw(output='mpl')
```

```python
result = get_result_from_backend(qc, 1000)
    
# Get the counts of the measurements
counts = result.get_counts()

# Plot the histogram
plot_histogram(counts)
```

# Let's match the matrices

Let's now learn a new thing, how to see the unitary matrix equivalent of a circuit. directly at Qiskit. And let's match with our knowledge for Pauli-X-Gate.

```python
from qiskit import BasicAer

# until now, we were using QASM simulator, now we need unitary simulator
backend = BasicAer.get_backend('unitary_simulator')
```

```python
q = QuantumRegister(1)

qcx = QuantumCircuit(q)
qcx.x(q)

backend.run(transpile(qcx,backend)).result().get_unitary(qcx, decimals=3)
```

So, it matches with what we learnt before. Let's test the matrices for Y and Z gate as well. Let's first define a function and call that subsequently.

```python
def get_unitary_matrix(qc):
    backend = BasicAer.get_backend('unitary_simulator')
    return backend.run(transpile(qc,backend)).result().get_unitary(qc, decimals=3)
```

```python
qcy = QuantumCircuit(q)
qcy.y(q)

print('Paui Y:\n', get_unitary_matrix(qcy)) # For Pauli-Y Gate

qcz = QuantumCircuit(q)
qcz.z(q)

print('Paui Z:\n',get_unitary_matrix(qcz)) # For Pauli-Y Gate
```

From the unitary matrix of Pauli-Z gate, please note an interesting thing, it will keep the $|0\rangle$ state unchanged and it will change $|1\rangle$ into -$|1\rangle$. Due to this, Pauli Z gate is called Phase Flip Gate.

Similarly, Pauli X-gate is called the bit flip gate as it swaps the coefficients of $|0\rangle$ and $|1\rangle$.

Just as expected, which proves, all the Pauli gates are their self-inverses.

# But Rotations Differ

We already know, that the axes about which Pauli gates rotate a qubit is different for X, Y and Z gates. Let's explore the same by visualizing it on the Bloch Sphere.

```python
# Create a quantum circuit with a single qubit
qc = QuantumCircuit(1)

# Apply Pauli X gate
qc.x(0)

# Visualize the circuit
print(qc)

# Simulate the circuit to get the final statevector
backend = Aer.get_backend('statevector_simulator')
compiled_circuit = transpile(qc, backend)
result = backend.run(compiled_circuit).result()
statevector = result.get_statevector()

# Visualize the Bloch sphere representation of the final state
plot_bloch_multivector(statevector)
```
